```python
#Import the libraries
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt   
```

```python
# Import .csv file and convert it to a DataFrame object
eastWest_data_xl = pd.ExcelFile("D:/datascience Excelr/class docs and assignments/Assignments/In progress/Clustering/EastWestAirlines/EastWestAirlines.xlsx")
  
```

```python
eastWest_data_xl.sheet_names
```

```python
eastWest_data=df1 = eastWest_data_xl.parse('data')
eastWest_data.head()
```

```python
eastWest_data1 = eastWest_data.drop(['ID#'],axis=1)   
```

```python
array=eastWest_data1.values
array  
```

```python
stscaler = StandardScaler().fit(array)
X = stscaler.transform(array) 
X  
```

```python
dbscan = DBSCAN(eps=0.45, min_samples=5)
dbscan.fit(X)                
```

```python
#Noisy samples are given the label -1.
dbscan.labels_     
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster']) 
```

```python
cl
pd.set_option("display.max_rows", None) 
```

```python
cl   
```

```python
eastWest_data1 = pd.concat([eastWest_data,cl],axis=1) 
eastWest_data1     
```

```python
import matplotlib.pyplot as plt
>>> plt.style.use('classic') 
```

```python
df1.plot(x="Balance",y ="Balance",c=dbscan.labels_ ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using DBScan')   
```

```python
dl = dbscan.labels_   
```

```python
import sklearn
sklearn.metrics.silhouette_score(X, dl) 
```

```python
from sklearn.cluster import KMeans
clf = KMeans(n_clusters=3)
y_kmeans = clf.fit_predict(X)  
```

```python
y_kmeans
```

```python
cl1=pd.DataFrame(y_kmeans,columns=['Kcluster']) 
cl1 
```

```python
df2 = pd.concat([df1,cl1],axis=1) 
df2.head()
```

```python
df2.plot(x="Balance",y ="Days_since_enroll",c=y_kmeans ,kind="scatter",s=50 ,cmap=plt.cm.copper_r) 
plt.title('Clusters using KMeans')   
```
